```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```


2. Dataset Description and Loading

The dataset used in this project contains structured data related to the problem domain. It includes multiple features that help in understanding trends, relationships, and outcomes. The dataset is loaded using the Pandas library for further processing and analysis.


```python
# Load the dataset
df = pd.read_csv("superstore.csv", encoding="latin1")

# Display first 5 rows
df.head()

```

## Data Cleaning and Preprocessing
The dataset is cleaned by handling missing values, removing duplicates, and converting data types where necessary.

```python
# Convert Order Date to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Check missing values
df.isnull().sum()
```

## Exploratory Data Analysis (EDA)
EDA is performed to understand sales and profit distribution across regions and categories.

```python
# Exploratory Data Analysis (EDA)

# Total sales and profit
total_sales = df['Sales'].sum()
total_profit = df['Profit'].sum()

print("Total Sales:", total_sales)
print("Total Profit:", total_profit)

print("\nSales Summary:")
print(df['Sales'].describe())

print("\nProfit Summary:")
print(df['Profit'].describe())
```

```python
# Group the dataset by 'Customer Name' and calculate total sales for each customer
top_customers = (
    df.groupby("Customer Name")["Sales"]
    .sum()                      # Calculate total sales per customer
    .sort_values(ascending=False)  # Sort customers by sales in descending order
    .head(5)                    # Select the top 5 customers with highest sales
)

# Display the top 5 customers by total sales
top_customers
```

## Visualizations

```python
# Group the data by 'Category' and calculate total Sales for each category
sales_by_category = df.groupby("Category")["Sales"].sum()

# Create a bar chart to visualize total sales by category
sales_by_category.plot(kind='bar', title="Sales by Category")

# Set label for x-axis
plt.xlabel("Category")

# Set label for y-axis
plt.ylabel("Sales")

# Display the plot
plt.show()

```

## Model Building and Evaluation
This task focuses on descriptive analytics and visualization; therefore, no predictive model is applied.

## Final Conclusion with Insights
The analysis shows regional and category-wise variations in sales and profit. Interactive dashboards help businesses track KPIs effectively and identify top-performing customers.
